# Dataframe.plot

```python
np.random.seed(123)
df = pd.DataFrame({'A':np.random.randn(365).cumsum(0), 'B':np.random.randn(365).cumsum(0)+20,'C':np.random.randn(365).cumsum(0)-20}, 
              index = pd.date_range('1-1-2017', periods=365))
df.head()
```

```python
plt.figure(figsize=(8,6))
df.plot();
```

```python
plt.scatter(df['A'], df['B'])
```

```python
plt.scatter(df['A'], df['C'], c=df['B'], s=df['B'], cmap='viridis')
```

```python
df.plot.box();
```

```python
df.plot.hist(alpha=0.8);
```

```python
df.plot.kde();
```

# Seaborn

```python
import seaborn as sns
```

```python
np.random.seed(1234)

v1 = pd.Series(np.random.normal(0,10,1000), name='v1')
v2 = pd.Series(2*v1 + np.random.normal(60,15,1000), name='v2')
```

```python
plt.figure()
plt.hist(v1, bins=np.arange(-50,150,5), label='v1')
plt.hist(v2, bins=np.arange(-50,150,5), label='v2')
plt.legend()
```

```python
# plot a kernel density estimation over a stacked barchart
plt.figure()
plt.hist([v1,v2],histtype='barstacked', density=True)
v3=np.concatenate((v1,v2))
sns.kdeplot(v3);
```

```python
sns.distplot(v3, hist_kws={'color':'Teal'}, kde_kws={'color':'Navy'});
```

```python
sns.jointplot(v1,v2, alpha=0.5)
```

```python
sns.jointplot(v1,v2, alpha=0.5, kind='hex')
```

```python
sns.jointplot(v1,v2, alpha=0.5, kind='kde',space=0)
```

```python
sns.pairplot(iris, diag_kind='kde', hue='Name')
```

```python
plt.figure(figsize=(8,6))
plt.subplot(121)
sns.swarmplot(iris['Name'], iris['PetalLength'])

plt.subplot(122)
sns.violinplot(iris['Name'], iris['PetalLength'])
```
