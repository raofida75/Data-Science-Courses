# Video 2

```python
df = pd.read_csv('../Data/starbucks.csv', index_col='Date', parse_dates=True)
```

```python
df.head()
```

```python
df.resample('A').mean()['Close']
# year end resampling
```

```python
df.resample('AS').mean()['Close']
# year start
```

```python
df.resample('W').mean()['Close']
# weekly
```

```python
df.resample('MS').mean()['Close']
# month start
```

```python
df.resample('M').mean()['Close'].plot.bar(figsize=(12,6))
```

# Video 3

```python
df.shift(1).head()
# all values are shifted 1 column down, lost record is lost
```

```python
df.shift(-1).head()
# all values are shifted 1 column up, first record is lost
```

```python
df.shift(periods=1, freq='M').head()
# all dates are shifted to the end of that month
```

```python
df.shift(periods=1, freq='Y').head()
# all dates are shifted to the end of that year
```

```python
df.head()
```

```python
df.plot(figsize=(10,5))
```

```python
for i in df.columns:
    df[i].plot(figsize=(10,5))
    plt.title(i)
    plt.show()
```

```python
df['Close'].plot(figsize=(10,6))
plt.ylabel('Closing Price')
plt.xlabel('DATE')
plt.title('Starbucks Closing Stock Prices')
plt.autoscale(axis='x', tight=True)
plt.show()
```

```python
# Plotting for 2017 only
```

```python
df.loc['2017-1-1': '2017-3-1']['Close'].plot(figsize=(10,5))
plt.autoscale(axis='x', tight=True)

# Choosing X Limits by Slice:
```

```python
df['Close'].plot(figsize=(10,6))
plt.xlim('2017-1-1', '2017-3-1')
plt.ylim(50,58);
# Choosing X Limits by Argument:
```

```python
title='Starbucks Closing Stock Prices'
ylabel='Closing Price (USD)'
xlabel='Closing Date'

ax = df['Close'].plot(xlim=['2017-01-04','2017-03-01'],ylim=[51,57],figsize=(12,4),title=title)
ax.set(xlabel=xlabel, ylabel=ylabel);

# Title and axis labels
```

```python
title='Starbucks Closing Stock Prices'
ylabel='Closing Price (USD)'
xlabel='Closing Date'

ax = df['Close'].plot(xlim=['2017-01-04','2017-03-01'],ylim=[51,57],figsize=(12,4),title=title, c='r', ls='--')
ax.set(xlabel=xlabel, ylabel=ylabel);
# Color and Style
```

```python
from matplotlib import dates
```

```python
ax = df.loc['2017-1-1': '2017-5-1']['Close'].plot(figsize=(12,5))
ax.xaxis.set_major_locator(dates.WeekdayLocator(byweekday=6))
ax.xaxis.set_major_formatter(dates.DateFormatter('%d'))

ax.xaxis.set_minor_locator(dates.MonthLocator())
ax.xaxis.set_minor_formatter(dates.DateFormatter('\n\n%b-%Y'))
ax.set(xlabel='')
ax.set(title='2017 Starbucks Closing Stock Prices')
plt.grid();

# Major vs. Minor Axis Values
```

```python
ax = df.loc['2015-1-1': '2017-12-31']['Close'].plot(figsize=(20,5))
ax.xaxis.set_major_locator(dates.MonthLocator())
ax.xaxis.set_major_formatter(dates.DateFormatter('%d-%b'))

ax.set(xlabel='')
ax.set(title='2015-2017 Starbucks Closing Stock Prices')
plt.autoscale(axis='x', tight=True)
plt.grid();
```
